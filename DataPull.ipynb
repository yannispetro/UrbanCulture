```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from random import randint
import numpy.random as random
from time import sleep
```

```python
class DataPull:
    def __init__(self):
        url = 'http://insideairbnb.com/get-the-data.html'
        res = requests.get(url)
        html_page = res.content

        soup = BeautifulSoup(html_page, 'html.parser')
        
        df0 = pd.DataFrame(columns = ['city','country','date','url'])
        for i, a in enumerate(soup.find_all('a', href=True, text = 'reviews.csv.gz')):
#         for i, a in enumerate(soup.find_all('a', href=True, text = 'listings.csv.gz')):
            urlComps = a['href'].split('/')
            df0.loc[i] = [urlComps[-4],urlComps[-6],urlComps[-3],a['href']]
            
        df0['date'] = pd.to_datetime(df0['date'])
        self.df_url = df0[df0.groupby(['city','country'])['date'].transform(max) == df0['date']]
        
    def cityDF(self,city):
        url = self.df_url[self.df_url['city']==city]['url'].values[0]
        self.df_city = pd.read_csv(url,compression='gzip')
        return self.df_city
    
    def downloadCSV(self,city):
        url = self.df_url[self.df_url['city']==city]['url'].values[0]
        r = requests.get(url, allow_redirects=True)
        open('files/' +  city + '_reviews.csv.gz', 'wb').write(r.content)
#         open('files/' +  city + '_listings.csv.gz', 'wb').write(r.content)
```

```python
dpull = DataPull()
```

```python
df_url = dpull.df_url
```

```python
df_url.to_csv('data/allCitiesData.csv.gz', compression='gzip')
```

```python
dpull = DataPull()
allCities = list(dpull.df_url[~dpull.df_url.city.isin(['new-york-city','toronto','montreal',
                                                       'south-aegean','crete','barcelona',
                                                      'london','paris','madrid','rome','athens','berlin',
                                                      'thessaloniki','munich'])].city)

allCititesRandom = random.permutation(allCities)
for city in allCititesRandom:
    print(city)
    dpull.downloadCSV(city)
    sleep(randint(3,10))
```
